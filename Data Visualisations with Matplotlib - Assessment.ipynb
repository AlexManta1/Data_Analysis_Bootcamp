```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/alexandermanta/Downloads/cirrhosis.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#1. What is the distribution of Albumin? (histogram)

df['Albumin'].plot(kind='hist');
```

```python
#2. What is the gender composition of respondents? (pie chart)

df['Sex'].value_counts().plot(kind='pie');
```

```python
#3. Is there a correlation between age and cholesterol?

correlation = df['Age'].corr(df['Cholesterol'])
correlation
```

```python
#4. Make a scatterplot of platelets vs. age. (you may also use regplot in seaborn)

df.plot(x='Platelets', y='Age',kind='scatter');
```

```python
#5. Modify the scatter plot above by adding the argument ‘Sex’ to the hue argument.

sns.pairplot(x_vars='Platelets', y_vars='Age',hue= 'Sex', kind='scatter',height = 3.5,data=df);
```

```python
#6. Make a bar graph or count plot of the Ascites column.

df['Ascites'].value_counts().plot(kind='bar');
```

```python
#7. Comment on the distribution of Prothrombin.

df['Prothrombin'].plot(kind='box');

df.Prothrombin.describe()

#As per the box plot represetation, the majority to values are within 5-13 range with a mean of 10.7. After 13.5 we can see slower distrubution of values until 18 and an outlier of 24.
```
