```python
#Creating Data Visualisations with Pandas and Matplotlib
```

```python
#Import the first package (Pandas)

import pandas as pd
```

```python
#Define the dataframe

df = pd.DataFrame({'year':[2010,2015,2020],'population':[2000,3400,7000]})
```

```python
#Verify the output

df.head()
```

```python
#Creating a line chart for the dataframe with .plot method

df.plot(x='year', y='population', kind='line');
```

```python
df = pd.DataFrame({'year':[2010,2015,2020],'population':[2000,3400,7000]})
```

```python
#Creating a bar chart for the dataframe with .plot method

df.plot(x='year', y= 'population', kind='bar');
```

```python
#Defining a new dataframe for a horizontal representation of the data

df = pd.DataFrame({'country':['UK','CAN','IND'],'population':[2000,3400,7000]})
```

```python
#Creating a bar chart for the new dataframe using .plot method

df.plot(x='country', y= 'population', kind='barh');
```

```python
dfbar = pd.DataFrame({'country':['UK','CAN','IND'],'population':[2000,3400,7000]})
```

```python
#Creating a pie chart for the new dataframe using .plot method and also defining the labels

dfbar.plot(labels=dfbar['country'], y= 'population', kind='pie');

#or

dfbar.plot(x = 'country', y = 'population', kind = 'pie', labels = ['UK', 'CAN', 'IND'])

#Cases where to use pie charts:
#1. When you have 5 or less categories
#2. When there is a clear visual distinction of categories
```

```python
#Defining a new dataframe for a scatter plot representation of the data

dfcor = pd.DataFrame({'eggs_eaten':[50000,60000,450592],'population':[2000,2200,2100]})
```

```python
#Creating a scatter plot for the new dataframe using .plot method and also defining the labels

dfcor.plot(x='eggs_eaten',y='population',kind='scatter');

#or

dfbar.plot(labels=dfbar["country"],y="population",kind="pie", figsize=(5,5));
```

```python
#Distribution

#Verifying the correlation between data
dfcor['population'].corr(dfcor['eggs_eaten'])
```

```python
#Creating a box plot using .plot method

dfcor.plot(y='population',kind='box');
```

```python
#Creating a histogram using .plot method on the Population Column from the dfcor dataframe


dfcor.plot(y='population',kind='hist', figsize=(4,3),title='Distribution', xlabel='range of pop',color='green');
```

```python
#Importing Matplotlib to create visualisations

import matplotlib.pyplot as plt
```

```python
dfcor = pd.DataFrame({'eggs_eaten':[50000,60000,450592],'population':[2000,2200,2100]})
```

```python
#Creating a barchart using .plot method and also defining chart characteristics

dfcor.plot(x='eggs_eaten', y= 'population', kind='bar')
plt.title('population vs eggs',y=1,x=1)
plt.xlabel('pop')
plt.ylabel('eggs')
```
